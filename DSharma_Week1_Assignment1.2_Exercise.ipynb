### 7. Find the five publishers with the highest total global sales. Note: You will need to calculate the total global sales for each publisher to do this.

```python
## Create dataframe with only 'Publisher', 'Global_Sales'.
## Calculate the sum of global sales and sort the data by 'Global_Sales' in descending order.
## Display the top 5 rows using the head function.

VGames_GS_df = Video_Games_df[['Publisher', 'Global_Sales']]
VGames_GS_sort_df = VGames_GS_df.groupby(['Publisher']).sum().sort_values(by='Global_Sales', ascending=False)
VGames_GS_sort_df.head(5)
```

### 10. Try to calculate the median user score of all the video games. You will likely run into an error because some of the user score entries are a non-numerical string that cannot be converted to a float. Find and replace this string with NaN and then calculate the median. Then, replace all NaN entries in the user score column with the median value.

```python
## Using median function on User_Score column, this will give us error due to non numerical string values.

Video_Games_df["User_Score"].median()
```

##### We can see the median function failed on Under_score column due to presence of non numerical values and error also suggest that string value cannot be converted to float. Let convert the non numerical values to NAN.

```python
#### Replacing the non numerical values to NAN.

VG_median_df = Video_Games_df.replace('tbd', np.nan)
VG_median_df.head(10)
```

```python
## Lets calculate the median now to see if it work.

median_vg = VG_median_df["User_Score"].median()
print("The median values for user score is : {}".format(median_vg))
```

```python
## Now we have median value 7.5. Lets replace the NAN values with median value 

VG_median_df['User_Score'].fillna(value=VG_median_df['User_Score'].median(), inplace=True)
VG_median_df.head(10)
```

```python
## Lets check if under_score column has any NAN values.

count_NAN = VG_median_df["User_Score"].isna().sum()
print("The number of NAN values in Under_Score column in dataframe is: {}".format(count_NAN))
```
